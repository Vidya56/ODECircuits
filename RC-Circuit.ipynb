### Solving circuits problem using numerical methods and DFT

### RC Circuit

```python
import numpy as np
import matplotlib.pyplot as plt

R = 5.0
C = 1.0
Vb = 1.5


t = np.linspace(0,10)
delta = 1

Vc = np.zeros(len(t))
I = np.zeros(len(t))
Vc[0] = 0
for i in range(1,len(Vc)):
    Vc[i] = Vc[i-1] + (delta/( R * C )) * (Vb - Vc[i-1])
    I[i-1]= C * (Vc[i]-Vc[i-1])

plt.plot(t,Vc)
plt.xlabel('Time (s)')
plt.ylabel('Vc(t)')
plt.title('Voltage across the capacitor')
plt.show()

plt.plot(t,I)
plt.xlabel('Time (s)')
plt.ylabel('I(t)')
plt.title('Current through the circuit')
plt.show()
```
